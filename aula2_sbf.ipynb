# Mecânica Quântica

## Postulados da Mecânica Quântica

### Postulado dos Estados
$$|\psi\rangle\in\mathcal{H} \text{ com } ||\psi||=1$$
Um qubit - Base computacional:
$$|0\rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix}, |1\rangle = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$$
Estado geral ($c_{j}\in\mathbb{C}$ para $j=0,1$ com $|c_0|^2 + |c_1|^2 =1$):
$$|\psi\rangle = \begin{bmatrix} c_{0} \\ c_{1} \end{bmatrix} = c_{0}\begin{bmatrix} 1 \\ 0 \end{bmatrix}+c_{1}\begin{bmatrix} 0 \\ 1 \end{bmatrix} = c_{0}|0\rangle+c_{1}|1\rangle$$

#### Sistemas compostos
Graus de liberdade $s=1,2,\cdots,n$:
$$|\Psi\rangle \in \mathcal{H} = \mathcal{H}_{1}\otimes\mathcal{H}_{2}\otimes\cdots\otimes\mathcal{H}_{n} \text{ com } ||\Psi||=1.$$

$$\text{Produto tensorial: }A\otimes B := \begin{bmatrix} A_{11}B & A_{12}B & \cdots \\
                    A_{21}B & A_{22}B & \cdots \\
                    \vdots & \vdots & \ddots\end{bmatrix}.$$                    

$$\text{Ex. }|01\rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix}\otimes\begin{bmatrix} 0 \\ 1 \end{bmatrix} = \begin{bmatrix} 0 \\ 1 \\ 0 \\ 0\end{bmatrix},\ |00\rangle=\begin{bmatrix} 1 \\ 0 \\ 0 \\ 0\end{bmatrix},\ |10\rangle=\begin{bmatrix} 0 \\ 0 \\ 1 \\ 0\end{bmatrix},\ |11\rangle=\begin{bmatrix} 0 \\ 0 \\ 0 \\ 1\end{bmatrix}$$
Notação: $|\psi\rangle\otimes|\phi\rangle = |\psi\rangle|\phi\rangle = |\psi\phi\rangle$

#### Estados emaranhados

Definição de emaranhamento: $|\Psi\rangle_{AB} \ne |\psi\rangle_A\otimes|\phi\rangle_B$ <br>

Base de Bell:
\begin{align}
& |\Phi_\pm\rangle = \frac{1}{\sqrt{2}}\big(|00\rangle\pm|11\rangle\big) = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 0 \\ 0 \\ \pm1\end{bmatrix},\\
& |\Psi_\pm\rangle = \frac{1}{\sqrt{2}}\big(|01\rangle\pm|10\rangle\big) = \frac{1}{\sqrt{2}}\begin{bmatrix} 0 \\ 1 \\ \pm1 \\ 0\end{bmatrix}.
\end{align}

### Postulado das medidas

Observáveis:
$$O\in L(\mathcal{H}) \text{, } O=O^{\dagger}=\sum_{j}o_j|o_j\rangle\langle o_j|$$
Regra de Born:
$$Pr(o_j|\psi)=|\langle o_j|\psi\rangle|^{2}$$
$\text{ Repetibilidade: }|\psi\rangle\xrightarrow[]{o_j}|o_j\rangle$

Exemplo:
$$\text{Prepara }|\psi\rangle = c_0|0\rangle+c_1|1\rangle \xrightarrow{\text{Mede}\ \{|0\rangle,|1\rangle\}}\begin{cases}\text{obtém }|0\rangle \text{ com prob. }|c_0|^2 \\ \text{obtém }|1\rangle \text{ com prob. }|c_1|^2\end{cases}$$

#### Postulado da dinâmica
$$|\psi_{t}\rangle=U_{t}|\psi_{0}\rangle \text{ com } i\hbar\partial_{t}U_{t}=H_{t}U_{t} \text{ e } UU^\dagger = U^\dagger U = I$$

Exemplos - __matrizes de Pauli__:
$$Z=\begin{bmatrix}1&0\\0&-1\end{bmatrix}\ \therefore Z|0\rangle=|0\rangle,\ Z|1\rangle=-|1\rangle$$
$$X=\begin{bmatrix}0&1\\1&0\end{bmatrix}=NOT\ \therefore X|0\rangle=|1\rangle,\ X|1\rangle=|0\rangle$$
$$Y=\begin{bmatrix}0&-i\\i&0\end{bmatrix}\ \therefore Y|0\rangle=i|1\rangle,\ Y|1\rangle=-i|0\rangle$$
Autoverores: $|\pm\rangle=(|0\rangle\pm|1\rangle)/\sqrt{2}$ pra $X$ e $|\oplus\rangle=(|0\rangle+i|1\rangle)/\sqrt{2}$ e $|\ominus\rangle=(|0\rangle-i|1\rangle)/\sqrt{2}$ pra $Y$

$$\text{Porta Hadamard: }H=\frac{1}{\sqrt{2}}\begin{bmatrix}1&1\\1&-1\end{bmatrix}\ \therefore H|0\rangle=|+\rangle,\ H|1\rangle=|-\rangle$$

$$\text{Porta Ctrl-NOT: }C_{X}^{A\rightarrow B} = |0\rangle\langle 0|\otimes\mathbb{I} + |1\rangle\langle 1|\otimes X\ \therefore\ C_{X}^{A\rightarrow B}|j\rangle|k\rangle=|j\rangle|j\oplus k\rangle$$

__Base computacional $\rightarrow$ Base de Bell__
$$|\psi_1\rangle = (H\otimes\mathbb{I})|00\rangle = |+\rangle\otimes|0\rangle = 2^{-1/2}(|00\rangle+|10\rangle) $$
Aplica a CNOT com o qubit coerente como controle
\begin{align}
|\Phi_+\rangle & = C_{X}^{A\rightarrow B}|\psi_1\rangle = 2^{-1/2}(|00\rangle + |11\rangle).
\end{align}
Outros estados da base de Bell ($U=C_{X}^{A\rightarrow B}(H\otimes I)$):
$$U|10\rangle=|\Phi_-\rangle,\ U|01\rangle=|\Psi_+\rangle,\ U|11\rangle=|\Psi_-\rangle$$
__Base de Bell $\rightarrow$ base computacional__ ($U^\dagger=(H\otimes I)C_{X}^{A\rightarrow B}$)
$$|00\rangle = U^\dagger|\Phi_+\rangle,\ |01\rangle = U^\dagger|\Psi_+\rangle,\ |10\rangle = U^\dagger|\Psi_-\rangle,\ |11\rangle = U^\dagger|\Phi_-\rangle.$$

__Medidas em uma base qualquer__ $$|b_{j_0 \cdots j_n}\rangle = U |j_0 \cdots j_n\rangle$$ a partir de medidas na base $|j_0 \cdots j_n\rangle$:
\begin{align}
& Pr(b_{j_0 \cdots j_n}|\psi) = |\langle b_{j_0 \cdots j_n}|\psi\rangle|^2 = |\langle j_0 \cdots j_n|U^\dagger|\psi\rangle|^2 \\
& \equiv |\langle j_0 \cdots j_n|\psi'\rangle|^2 = Pr(j_0 \cdots j_n|\psi')
\end{align}
em que
$$|\psi'\rangle := U^\dagger|\psi\rangle.$$

__Medidas na base $\{|+\rangle,|-\rangle\} = \{H|0\rangle,H|1\rangle\}$__ de um qubit
$$Pr(+|\psi) = Pr(0|\psi')$$
$$Pr(-|\psi)=Pr(1|\psi'),$$
com $$|\psi'\rangle=H^\dagger|\psi\rangle=H|\psi\rangle.$$

$$\text{ Porta } S = \begin{bmatrix}1&0\\0&i\end{bmatrix}\ \therefore\ S^{\dagger} = \begin{bmatrix}1&0\\0&-i\end{bmatrix}\ \therefore\ S|0\rangle=|0\rangle, S|1\rangle=i|1\rangle = e^{i\pi/2}|1\rangle$$

$$\text{ Porta de fase: } P(\theta) = \begin{bmatrix}1&0\\0&e^{i\theta}\end{bmatrix}\ \therefore\ P(\theta)|0\rangle=|0\rangle, P(\theta)|1\rangle = e^{i\theta}|1\rangle$$

__Medidas na base $\{|\oplus\rangle,|\ominus\rangle\}$__
\begin{align}
& S|+\rangle = (|0\rangle+i|1\rangle)/\sqrt{2} = |\oplus\rangle = SH|0\rangle \\
& S|-\rangle = (|0\rangle-i|1\rangle)/\sqrt{2} = |\ominus\rangle = SH|1\rangle.
\end{align}
Assim
\begin{align}
& Pr(\oplus|\psi) = Pr(0|\psi'), \\
& Pr(\ominus|\psi)=Pr(1|\psi')
\end{align}
com 
$$|\psi'\rangle = (SH)^\dagger|\psi\rangle = HS^{\dagger}|\psi\rangle.$$

__Observável geral__ de um sistema de 2 níveis
$$\sigma_n = \hat{n}\cdot\vec{\sigma} = \begin{bmatrix}
\cos\theta & e^{-i\phi}\sin\theta \\ e^{i\phi}\sin\theta & -\cos\theta \\ \end{bmatrix}$$
aplica
$$|\psi'\rangle = (U(\theta,\phi,\pi-\phi))^{\dagger}|\psi\rangle = U(\theta,\phi,\pi-\phi)|\psi\rangle$$
Porta de rotação geral de 1 qubit
$$U(\theta,\phi,\lambda) = \begin{bmatrix} \cos(\theta/2) & -e^{i\lambda}\sin(\theta/2) \\ e^{i\phi}\sin(\theta/2) & e^{i(\phi+\lambda)}\cos(\theta/2) \end{bmatrix}$$
Assim, e.g.,
\begin{align}
& U|0\rangle = \cos(\theta/2)|0\rangle + e^{i\phi}\sin(\theta/2)|1\rangle, \\
& U|1\rangle = -e^{i\lambda}\sin(\theta/2)|0\rangle + e^{i(\phi+\lambda)}\cos(\theta/2)|1\rangle.
    \end{align}

### Tomografia de estados

__Postulado dos estados:__ $\rho:\mathcal{H}\rightarrow\mathcal{H}$ com $\rho\ge 0$ e $Tr(\rho)=1$ <br>
__Postulado das medidas:__ $Pr(o_j|\rho) = Tr(\rho|o_j\rangle\langle o_j|)$.

$*$ Estados de __1 qubit:__
$$\rho = (1/2)\big(I + \langle X\rangle_\rho X + \langle Y\rangle_\rho Y +\langle Z\rangle_\rho Z\big),$$
com e.g. $\langle X\rangle_\rho = Tr(\rho X) = (+1)Pr(x_+|\rho)+(-1)Pr(x_-|\rho)$.

$*$ Estados de __2 qubits:__
$\rho = \frac{1}{4}\sum_{j,k=0}^3 \langle \sigma_j\otimes\sigma_k\rangle_\rho  \sigma_j\otimes\sigma_k,$
com $\sigma_0 = I,\ \sigma_1=X,\ \sigma_2=Y,\ \sigma_3=Z$ e e.g. 
\begin{align}
\langle X\otimes Y\rangle = &(+1)(+1)Pr(x_+,y_+)+(+1)(-1)Pr(x_+,y_-)\\
& +(-1)(+1)Pr(x_-,y_+)+(-1)(-1)Pr(x_-,y_-).
\end{align}

## Tomografia de estados no qiskit

```python
# Cria o circuito quântico, sem usar medidas

from qiskit import *
qr = QuantumRegister(1)
qc = QuantumCircuit(qr)
qc.h(qr[0])
qc.s(qr[0])
qc.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator') 

nshots = 8192
```

```python
import qiskit
from qiskit.ignis.verification.tomography import state_tomography_circuits, StateTomographyFitter
qstc = state_tomography_circuits(qc, [qr[0]])
job = qiskit.execute(qstc, simulator, shots=nshots)
qstf = StateTomographyFitter(job.result(), qstc)
rho = qstf.fit(method='lstsq')
qiskit.visualization.plot_state_city(rho)
```

# O paradoxo EPR

A. Einstein, B. Podolsky, and N. Rosen, “Can Quantum-Mechanical Description of Physical Reality Be Considered Complete?,” Phys. Rev. 47, 777 (1935).

__Teoria completa__: Cada elemento da teoria corresponde a um elemento de realidade. <br>
__Realismo__: Uma quantidade física é real se podemos prevê-la com certeza sem perturbar o sistema.

Argumento: Consideremos o seguinte estado de spin de dois quantons de spin 1/2:
\begin{equation}
|\Psi_{-}\rangle_{AB} = \frac{1}{\sqrt{2}}\left(|0\rangle_{A}\otimes|1\rangle_{B} - |1\rangle_{A}\otimes|0\rangle_{B}\right).
\end{equation}
Usando 
$$|\pm\rangle_{S}=(|0\rangle_{S}\pm|1\rangle_{S})/\sqrt{2},$$
para $S=A,B$, pode-se verificar que:
\begin{equation}
|\Psi_{-}\rangle_{AB} = \frac{e^{i\pi}}{\sqrt{2}}\left(|+\rangle_{A}\otimes|-\rangle_{B} - |-\rangle_{A}\otimes|+\rangle_{B}\right).
\end{equation}

Agora considere que medimos um dos observáveis $Z^{A}$ ou $X^{A}$, do quanton da Alice. Se $\{|0\rangle,|1\rangle\}^{A}$ ou $\{|+\rangle,|-\rangle\}^{A}$ é obtido nessas medidas da Alice, pela correlação que há entre os spins, sabemos que o estado do quanton do Bob depois das medidas será $\{|1\rangle,|0\rangle\}^{A}$ ou $\{|-\rangle,|+\rangle\}^{A}$. Como exemplo, consideremos
\begin{align}
|\Psi_{-}\rangle_{AB} \xrightarrow[]{|0\rangle_A} &\propto \Pi_{0}^{A}\otimes\mathbb{I}_{B}|\Psi_{-}\rangle_{AB} \\
& = |0\rangle_{A}\langle 0|\otimes\mathbb{I}_{B}|\Psi_{-}\rangle_{AB} \\
& = \frac{1}{\sqrt{2}}\left(|0\rangle_{A}\langle 0||0\rangle_{A}\otimes\mathbb{I}_{B}|1\rangle_{B} - |0\rangle_{A}\langle 0||1\rangle_{A}\otimes\mathbb{I}_{B}|0\rangle_{B}\right) \\
& = \frac{1}{\sqrt{2}}|0\rangle_{A}\otimes|1\rangle_{B}
\end{align}

e
\begin{align}
|\Psi_{-}\rangle_{AB} \xrightarrow[]{|+\rangle_{A}} & \propto \Pi_{+}^{A}\otimes\mathbb{I}_{B}|\Psi_{-}\rangle_{AB} \\
& = |+\rangle_{A}\langle +|\otimes\mathbb{I}_{B}|\Psi_{-}\rangle_{AB} \\
& = \frac{e^{i\pi}}{\sqrt{2}}\left(|+\rangle_{A}\langle +||+\rangle_{A}\otimes\mathbb{I}_{B}|-\rangle_{B} - |+\rangle_{A}\langle +||-\rangle_{A}\otimes\mathbb{I}_{B}|+\rangle_{B}\right) \\
& = \frac{e^{i\pi}}{\sqrt{2}}|+\rangle_{A}\otimes|-\rangle_{B}.
\end{align}

Localidade: As medidas feitas por Alice não podem afetar o resultado das medidas que foram, que serão ou que estão sendo feitas por Bob. <br>
Usando $|\Psi_-\rangle_{AB}$ podemos prever com certeza e sem perturbar o sistema $B$, os resultados de medidas de dois _observáveis incompatíveis_ $Z^{B}$ ou $X^{B}$ do quanton do Bob.

# Comunicação superluminal?
Esses resultados motivaram propostas engenhosas para fazer comunicação mais rápido que a luz. Uma proposta foi a seguinte. Considere que duas partes, chamadas usualmente de Alice e Bob, querem trocar bits de informação e que eles compartilham pares de quantons de spin 1/2 preparados no estado $|\Psi_{-}\rangle_{AB}$. Alice e Bob combinam que se Alice quer mandar o bit $0$ ela mede o observável $\sigma_{z}$ no seu spin do par e se quiser mandar o bit $1$ ela mede $\sigma_{x}$:
\begin{align}
& \text{ Alice mede } \sigma_{z}^{A} \rightarrow \text{ bit } 0, \\
& \text{ Alice mede } \sigma_{x}^{A} \rightarrow \text{ bit } 1.
\end{align}
Por sua vez, Bob faz várias cópias do estado do seu spin de cada par e mede o observável $\sigma_{z}^{B}$ em cada cópia:
\begin{equation}
|\psi\rangle_{B} \rightarrow \text{ xerox quântico } \rightarrow  |\psi\rangle_{B}\otimes|\psi\rangle_{B}\otimes\cdots \rightarrow\text{ mede }\sigma_{z}^{B}.
\end{equation}
Se Alice mediu $\sigma_{z}^{A}$, a distribuição de probabilidades obtida nas medidas de Bob, 
$$\{Pr(z_{+}^{B}),Pr(z_{-}^{B})\}=\{1,0\} \text{ ou }\{Pr(z_{+}^{B}),Pr(z_{-}^{B})\}\{0,1\},$$
pois
\begin{align}
\text{Alice obteve }z_{+}^{A} \Rightarrow \text{ estado colapsa pra } |z_{+}^{A}\rangle\otimes|z_{-}^{B}\rangle \Rightarrow \{0,1\}, \\
\text{Alice obteve }z_{-}^{A} \Rightarrow \text{ estado colapsa pra } |z_{-}^{A}\rangle\otimes|z_{+}^{B}\rangle\Rightarrow \{1,0\}.
\end{align}
Se Alice, no entanto, mediu $\sigma_{x}$, Bob obterá, aproximadamente, a DP 
$$\{Pr(z_{+}^{B}),Pr(z_{-}^{B})\}=\{1/2,1/2\}$$
para medidas de $\sigma_{z}^{B}$. Com isso, Bob consegue diferenciar entre as DPs e saber qual bit Alice quiz enviar pra ele. Se a distância entre eles for adequada, eles conseguem trocar bits de informação mais rápido do que a luz demoraria para ir da posição de Alice até Bob. Ou seja, eles fazem comunicação superluminal, voltam no tempo, e .... <br>
Qual é o problema com essa proposta?

# Teorema da não clonagem em MQ
O problema com a proposta anterior é que não existe Xerox quântico. Uma prova simplificada disto é a seguinte. Considera que existe uma dinâmica dada pelo operador de evolução temporal $U$ que atua no espaço $\mathcal{H}_{s}\otimes\mathcal{H}_{a}$, com $s$ sendo o sistema cujo estado queremos copiar e $a$ é um sistema auxiliar no qual  copiaremos aquele estado. Consideremos que $U$ pode copiar dois estados desconhecidos quaisquer $|\psi\rangle$ e $|\phi\rangle$, i.e.,
\begin{align}
U|\psi\rangle_{s}\otimes|\alpha\rangle_{a} & = |\psi\rangle_{s}\otimes|\psi\rangle_{a}, \\
U|\phi\rangle_{s}\otimes|\alpha\rangle_{a} & = |\phi\rangle_{s}\otimes|\phi\rangle_{a},
\end{align}
com $|\alpha\rangle$ sendo o estado inicial do sistema auxiliar. Segue dessas relações que
\begin{align}
\langle\psi|\otimes\langle\alpha|U^{\dagger}U|\phi\rangle\otimes|\alpha\rangle & = (\langle\psi|\otimes\langle\psi|)(|\phi\rangle\otimes|\phi\rangle) \\
\langle\psi|\phi\rangle\otimes\langle\alpha|\alpha\rangle & = \langle\psi|\phi\rangle\otimes\langle\psi|\phi\rangle \\
\langle\psi|\phi\rangle & = \langle\psi|\phi\rangle^{2}.
\end{align}
Ou seja, $\langle\psi|\phi\rangle$ é igual a $0$ ou $1$. Ou seja, só poderá existir uma máquina de copiar estados desconhecidos $U$ se os estados forem iguais ou ortogonais (que é o caso da informação clássica).

Uma transformação unitária que copia alguns estados é a chamada "não controlada":
\begin{equation}
CNOT_{s\rightarrow a} = |z_{+}\rangle_{s}\langle z_{+}|\otimes\mathbb{I}_{a}+|z_{-}\rangle_{s}\langle z_{-}|\otimes\sigma_{x}^{a}.
\end{equation}
Pode-se verificar que
\begin{align}
CNOT_{s\rightarrow a}|z_{+}\rangle_{s}\otimes|z_{+}\rangle_{a} & = |z_{+}\rangle_{s}\langle z_{+}||z_{+}\rangle_{s}\otimes\mathbb{I}_{a}|z_{+}\rangle_{a} + |z_{-}\rangle_{s}\langle z_{-}||z_{+}\rangle_{s}\otimes\sigma_{x}^{a}|z_{+}\rangle_{a} \\
& = |z_{+}\rangle_{s}\otimes|z_{+}\rangle_{a}, \\
CNOT_{s\rightarrow a}|z_{-}\rangle_{s}\otimes|z_{+}\rangle_{a} &= |z_{-}\rangle_{s}\otimes|z_{-}\rangle_{a}.
\end{align}
No entanto, para $c_{\pm}\in\mathbb{C}$ e
\begin{equation}
|\psi\rangle_{s}=c_{+}|z_{+}\rangle_{s}+c_{-}|z_{-}\rangle_{s},
\end{equation}
teremos 
\begin{align}
CNOT_{s\rightarrow a}|\psi\rangle_{s}\otimes|z_{+}\rangle_{a} & = c_{+}|z_{+}\rangle_{s}\otimes|z_{+}\rangle_{a}+c_{-}|z_{-}\rangle_{s}\otimes|z_{-}\rangle_{a} \\
& \ne|\psi\rangle_{s}\otimes|\psi\rangle_{a}
\end{align}
em geral.

Vale observar que máquinas de copiar estados quânticos de forma imperfeita ou probabilística são factíveis, mas não são "boas o suficiente" para fazer comunicação superluminal (https://arxiv.org/abs/quant-ph/0511088).
