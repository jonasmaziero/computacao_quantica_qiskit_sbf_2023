# A bomba

- Bomba não explodida: $|0\rangle;$
- Bomba explodida: $|1\rangle.$

A bomba é sensível a um único fóton, e ao absorvê-lo a bomba explodirá:
$$U_b|0\rangle=|1\rangle\ \therefore\ U_b=X.$$

Estado inicial da bomba: $|0\rangle$.

```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(1,1)
qc.x(0)
qc.measure(0,0)
qc.draw('mpl')
```

```python
from qiskit import Aer, execute
simulator = Aer.get_backend('qasm_simulator')
nshots = 2**13
job = execute(qc, backend=simulator,shots=nshots)
counts = job.result().get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

Ou seja, a bomba tem 100% de chance de explodir.

## Detectando a bomba sem ela explodir

Coloca a bomba em um dos caminhos de um interferômetro de Mach-Zehnder.

<table>
    <td><img src = 'fig_bomba.png' width = '300'></td>
    <td>Sem a bomba: $$Pr(D_0)=1,\ Pr(D_1)=0$$ </td>
    </table>

Ação da bomba (quântica): $$U_b = |0\rangle\langle 0|_C\otimes I_B + |1\rangle\langle 1|_C\otimes X_B$$

Sequência de estados
\begin{align}
\text{Inicial: } & |00\rangle_{CB}, \\
\text{DF: } & 2^{-1/2}(|0\rangle_C + i|1\rangle_C)|0\rangle_B, \\
\text{espelhos: } & 2^{-1/2}(i|1\rangle_C + ii|0\rangle_C)|0\rangle_B = 2^{-1/2}(i|10\rangle_{CB} - |00\rangle_{CB}), \\
\text{Bomba: } & 2^{-1/2}(i|11\rangle_{CB} - |00\rangle_{CB}), \\
\text{DF: } & 2^{-1}\big(i(|1\rangle_C + i|0\rangle_C)|1\rangle_{B} - (|0\rangle_C + i|1\rangle_C)|0\rangle_{B}\big) \\
& = 2^{-1}\big(i|11\rangle_{CB} - |01\rangle_{CB} - |00\rangle_{CB} - i|10\rangle_{CB}\big) 
\end{align}


Probabilidade da bomba explodir
\begin{align}
Pr(B=1) & = Pr(C=0,B=1) + Pr(C=1,B=1) \\
& = \frac{1}{4}+\frac{1}{4} = \frac{1}{2}.
\end{align}
Ou seja, a probabilidade da bomda explodir diminui pela metade! Até aí nenhuma novidade.

### Detectando a bomba sem explodir a bomba:
Sem a bomba temos $Pr(C=1)=0$. Portanto, so chega fóton no detector $D_1$ se a bomba estiver lá no caminho vertical. Assim, a chance de detectar a bomba sem que ela exploda é:
$$Pr(C=1,B=0) = \frac{1}{4}.$$

```python
from qiskit import QuantumCircuit

def qc_df(): # divisor de feixes 50:50
    qc = QuantumCircuit(1, name='DF')
    qc.s(0)
    qc.h(0)
    qc.s(0)
    return qc
qcdf = qc_df(); qcdf.draw('mpl')
```

```python
def qc_espelhos():
    qc = QuantumCircuit(1, name='E')
    qc.z(0)
    qc.y(0)
    return qc
qce = qc_espelhos(); qce.draw('mpl')
```

```python
def qc_bomba():
    qc = QuantumCircuit(2, name='B')
    qc.cx(0,1)
    return qc
qcb = qc_bomba(); qcb.draw('mpl')
```

```python
def qc_detector_de_bomba():
    qc = QuantumCircuit(2, name='DB')
    qcdf = qc_df(); qc.append(qcdf,[0])
    qce = qc_espelhos(); qc.append(qce,[0])
    qcb = qc_bomba(); qc.append(qcb,[0,1])
    qcdf = qc_df(); qc.append(qcdf,[0])
    return qc
qcdb = qc_detector_de_bomba(); qcdb.draw('mpl')
```

```python
qc = QuantumCircuit(2,2)
qcdb = qc_detector_de_bomba(); qc.append(qcdb,[0,1])
qc.measure([0,1],[0,1])
qc.draw('mpl')
```

```python
# Simulação clássica
from qiskit import Aer, execute
simulator = Aer.get_backend('qasm_simulator')
nshots = 2**13
job = execute(qc, backend=simulator,shots=nshots)
counts = job.result().get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
